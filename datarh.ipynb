```python
import folium

# coordenadas geográficas dos estados
coordenadas = {
    'SP': (-23.55, -46.64),
    'RJ': (-22.84, -43.15),
    'MG': (-18.10, -44.38),
    'GO': (-16.64, -49.31),
    'PA': (-5.53, -52.29),
    'CE': (-3.71, -38.54)
}

# cria um mapa centrado no Brasil
mapa = folium.Map(location=[-15.788497,-47.879873], zoom_start=4)

# adiciona um marcador para cada estado
for estado, coord in coordenadas.items():
    folium.Marker(location=coord, tooltip=estado).add_to(mapa)

# exibe o mapa no Jupyter Notebook
mapa
```

```python
print(df['NivelSatisfacao'].unique())
```

```python
def classificar_engajamento(nota):
    if nota > 4:
        return 'Alta'
    elif nota > 2:
        return 'Média'
    else:
        return 'Baixa'

df['Engajamento'] = df['NivelSatisfacao'].apply(classificar_engajamento)
```

```python
df.head(5)
```

```python
import matplotlib.pyplot as plt

plt.violinplot([df[df['Engajamento'] == 'Baixa']['Salario'],
             df[df['Engajamento'] == 'Média']['Salario'],
             df[df['Engajamento'] == 'Alta']['Salario']])
plt.xticks([1, 2, 3], ['Baixa', 'Média', 'Alta'])
plt.xlabel('Engajamento')
plt.ylabel('Salário')
plt.title('Salário em relação ao engajamento')
plt.show()
```
